```python
import pandas as pd

# Assuming df is your DataFrame containing the data
# Define the numeric columns
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Initialize an empty dictionary to store columns with outliers
outliers_dict = {}

# Calculate IQR for each numeric column
for col in numeric_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define the range to filter outliers
    filter = (df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))
    
    # Store columns with outliers in the dictionary
    if filter.any():
        outliers_dict[col] = df[col][filter].count()

# Print columns with outliers and the number of outliers
for col, count in outliers_dict.items():
    print(f"Column '{col}' has {count} outliers.")
```

```python
# Cap outliers
for col in columns_to_visualize:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define the range to cap outliers
    cap_min = Q1 - 1.5 * IQR
    cap_max = Q3 + 1.5 * IQR
    
    # Cap outliers
    df[col] = df[col].apply(lambda x: cap_min if x < cap_min else (cap_max if x > cap_max else x))
```
